# Kaggle Datasets

以下のリンクからKaggle データセットにアクセスできます:
> https://www.kaggle.com/datasets

リンクをクリックすると、独自のプロジェクトで使用できるデータセットのコレクションを含む Web ページが表示されます。

多くの新しいデータセットが毎日アップロードされるため、ページに表示されるデータセットは、上のスクリーンショットに示されているものとは異なる可能性があることに注意してください! <br>
Kaggle データセットには、CSV ファイルを含むさまざまなファイル タイプがありますが、JSON、SQLite、BigQuery などのよりエキゾチックなファイル タイプもあります。 <br>
少なくとも 1 つの CSV ファイルを含むデータセットを慎重に選択します。これは、このコースで扱ってきたファイル タイプであるためです。

特定のデータセットを検索するには、画面上部の検索バーを使用します。 たとえば、漫画のキャラクターに関するデータセットを操作したいとします。 検索ウィンドウに **"comic"** と入力して開始します。

すると、こちらが見つかります。[**FiveThirtyEight Comic Characters Dataset**](https://www.kaggle.com/fivethirtyeight/fivethirtyeight-comic-characters-dataset). <br>
データセットには、使用できる CSV ファイルを含む 3 つのファイルが含まれていることに注意してください。

次に、データセットをクリックして選択します。 これにより、データセットを説明する Web ページが表示されます。

![tut6_sample_dataset](https://i.imgur.com/4W1Sa7H.png)

下にスクロールして、ウィンドウの左側にある **Data Explorer** の下にあるデータセット内のファイルのリストを表示します。 データセットには、(1) **README.md**、(2) **dc-wikia-data.csv**、および (3) **marvel-wikia-data.csv** の 3 つのファイルが含まれています。 最初のファイルがデフォルトとして選択されます。 代わりに、CSV ファイルの 1 つをクリックして、ファイルのクイック プレビューを表示します。

このページの他のタブを調べてみてください。 たとえば、**ディスカッション** タブをチェックして、データセットについて他の人が何を言っているかを確認してください。

# あなた自身のデータセットを使う

次の演習では、選択した任意の CSV データセットを操作します。 上記で学んだように、Kaggle Datasets には、使用できるデータセットの大規模なコレクションが含まれています。 独自のデータセットで作業したい場合は、最初にそれを Kaggle Datasets にアップロードする必要があります。

こちらに戻り、[kaggle.com/datasets](https://www.kaggle.com/datasets), 画面左上にある"New Dataset"ボタンをクリックします。 

CSV ファイルの種類に慣れていない場合は、Kaggle Datasets プラットフォームが自動的に表形式のデータを CSV ファイルに変換することに注意してください。<br>
Google スプレッドシートや Excel ワークシートなどを自由にアップロードしてください。CSV ファイルに変換されます。
