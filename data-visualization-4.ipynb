# Set up the notebook

いつものように、Notebookを初期化します。

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

# Load and examine the data

今回は保険料のデータセットを使用します。<br>
一部の顧客が他の顧客よりも多く支払う理由を理解できるかどうかを確認します。

このデータセットの詳細は [こちら](https://www.kaggle.com/mirichoi0218/insurance/home) で確認できます。

```python
# Path of the file to read
insurance_filepath = "../input/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

データセットが正しく読み取れたか確認するために、先頭５行を出力してみます。

```python
insurance_data.head()
```

# Scatter plots

散布図を作成するために、`sns.scatterplot`を利用します:
- 水平方向 x軸 (`x=insurance_data['bmi']`)
- 垂直方向 y軸 (`y=insurance_data['charges`])

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

上の散布図は、[体格指数](https://en.wikipedia.org/wiki/Body_mass_index) (BMI) と保険料が **正の相関**を示しており、BMI が高い顧客は通常、より多く保険料を支払う傾向があります。 保険料で。 (_BMI が高いと一般的に慢性疾患のリスクが高くなるため、このパターンは理にかなっています._)

この関係の強さを再確認するには、**回帰線**、またはデータに最も適合する線を追加することをお勧めします。 これを行うには、コマンドを「sns.regplot」に変更します。

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

# Color-coded scatter plots

散布図を使用して、(_2つではなく..._) 3 つの変数間の関係を表示できます。 これを行う 1 つの方法は、点を色分けすることです。

たとえば、喫煙が BMI と保険料の関係にどのように影響するかを理解するために、ポイントを「喫煙者」別に色分けし、他の 2 つの列 (「bmi」、「料金」) を軸上にプロットできます。

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

この散布図は、非喫煙者は BMI が増加するとわずかに多く支払う傾向がある一方で、喫煙者ははるかに多く支払うことを示しています.

この事実をさらに強調するために、「sns.lmplot」コマンドを使用して、喫煙者と非喫煙者に対応する 2 つの回帰直線を追加できます。 (_喫煙者の回帰直線は、非喫煙者の回帰直線に比べて傾きが急であることがわかります!_)

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

この `sns.lmplot` コマンドは、これまでに学習したコマンドとは少し異なります。
- `x=insurance_data['bmi']`を設定する代わりに、`insurance_data`の`bmi`カラムを選択します。`x="bmi"`はカラム名のみ指定します。
- 同様に`y=charges` や`hue="smoker"`はカラム名を指定します。
- データセットの指定は`data=insurance_data`で指定します。

最後に、もう 1 つのプロットについて学習します。<br>
これは、見慣れている散布図とは少し異なります。 <br>
通常、散布図を使用して、2 つの連続変数 (`"bmi"` と `"charges"` など) の間の関係を強調します。 ただし、散布図のデザインを調整して、主軸の 1 つにカテゴリ変数 (「喫煙者」など) を配置することができます。<br>
このプロット タイプを **categorical scatter plot** と呼び、`sns.swarmplot` コマンドで作成します。

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

とりわけ、このプロットは次のことを示しています。
- 平均して、非喫煙者は喫煙者よりも安く請求されます。
- 最もお金を払う顧客は喫煙者です。 一方、最も支払いが少ない顧客は非喫煙者です。
